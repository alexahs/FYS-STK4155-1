## Exercise 2

```python
# Exercise 2.1: Own code for making polynomial fit
import numpy as np
import matplotlib.pyplot as plt

N = 100 # data size
p = 2   # polynomial degree

# Random data
x = np.random.rand(N, 1)
y = 5*x*x + 0.1*np.random.randn(N, 1)

# Creating design matrix X
X = np.ones((N, p + 1))
for i in range(1, p + 1):
    X[:,i] = x[:,0]**i
    
# Matrix inversion
beta = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)
# Creating model
y_tilde = X @ beta

# Sorting arrays for plotting
idx_sorted = np.argsort(x[:,0])

# Plotting model and data
plt.figure()
plt.plot(x, y, '.', label='random data')
plt.plot(x[idx_sorted], y_tilde[idx_sorted], '-', label='polynomial fit of degree 2')
plt.legend()
plt.show()
```

```python
# Exercise 2.2: Using scikit-learn to make polynomial fit
import sklearn.linear_model as skl
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

clf = skl.LinearRegression().fit(X, y)
y_tilde_scikit = clf.predict(X)

# Plotting model and data
plt.figure()
plt.plot(x, y, '.', label='random data')
plt.plot(x[idx_sorted], y_tilde_scikit[idx_sorted], 'r-', label='scikit polynomial fit of degree 2')
plt.legend()
plt.show()
```

```python
# Exercise 2.3: Error analysis                               
print("Mean squared error: %.2f" % mean_squared_error(y, y_tilde_scikit))
print('R2 score: %.2f' % r2_score(y, y_tilde_scikit))
print('Mean absolute error: %.2f' % mean_absolute_error(y, y_tilde_scikit))
```
